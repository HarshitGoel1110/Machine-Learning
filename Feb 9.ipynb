```python
import pandas as pd
import numpy as np
```

```python
# load data
data = pd.read_csv('crx.data' , header = None)
```

```python
data
```

```python
varnames = ['A' + str(c) for c in  range(1,17)]
varnames
```

```python
data.columns = varnames
```

```python
data
```

```python
# replace the ? with np.nan
data = data.replace('?' , np.nan)
data
```

```python
# re-cast some variables to its correct types
data.info()
data['A2'] = data['A2'].astype('float')
data.info()
```

```python
# encode target to binary
data['A16'] = data['A16'].map({'+' : 1 , '-' : 0})
data
# data.info()
```

```python
data.head(15)
```

```python
# save the data as csv
data.to_csv('now.csv' , index = False)
```

```python
# find categorical variables
# 'O' -> object or strings
# if non categorical
# data[c].dtype != 'O'
cat_columns = [c for c in data.columns if data[c].dtypes == 'O']
```
