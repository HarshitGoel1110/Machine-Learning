```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('lab_assess.csv')
data.head()
```

```python
data['Area'] = (data['Area'] - data['Area'].min()) / (data['Area'].max() - data['Area'].min())
data.head()
# standardise the variable Area
```

```python
data['Transaction_type'] = data['Transaction_type'].map({'Resale' : 0 , 'New Property' : 1})
data.head()
# encoding the variable Transaction_type
```

```python
data['Furnished_status'] = data['Furnished_status'].map({'Unfurnished' : 0 , 'SemiFurnished' : 1 , 'Furnished' : 2})
data
# encoding the variable Furnished_status
```

```python
data['City'] = data['City'].map({'Manali' : 0 , 'Shimla' : 1})
data
# encoding the variable City
```

```python
data['BHK'] = (data['BHK'] - data['BHK'].min()) / (data['BHK'].max() - data['BHK'].min())
data
# standardise the variable Area
```
