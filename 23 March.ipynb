```python
# Building a Decision Tree Classifier
#1 load the dataset
import pandas as pd
import numpy as np
```

```python
bank_df = pd.read_csv("bank_kaggle.csv")
bank_df.head()
```

```python
# #2 Work with X(Predictor Variables)
# Y = bank_df['deposit']
# X = bank_df.drop('deposit' , axis = 1)
```

```python
# X.head()
```

```python
# Y.head()
```

```python
#2 Work with X(Predictor Variables)
X_features = list(bank_df.columns)
X_features.remove('deposit')
X_features
```

```python
#3 Encoding for all the categorical variables for all the variable in one shot
encode_bank_df = pd.get_dummies(bank_df[X_features])
```

```python
encode_bank_df.head()
```

```python
#4 Set the X and Y
x = encode_bank_df
y = bank_df.deposit.map(lambda x: int(x == 'yes'))
```

```python
x.head()
```

```python
y.head()
```

```python
#5 Split the dataset
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.3 , random_state = 100)
```

```python
#6 Build the Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

# build the decision classfier based on "gini" criteria or the "entropy" criteria
clf_tree = DecisionTreeClassifier(criterion = 'gini' , max_depth = 3)
```

```python
clf_tree.fit(x_train , y_train)
```

```python
#7 Accuracy score
clf_tree.score(x_test , y_test)
```
